# Concatenation and Merging of two dataframes

```python
students = pd.DataFrame({'id_number' : [204, 202, 264, 136, 238],
                               'Name' : ['James Smith',
                                        'Maria Rodriguez',
                                        'David Smith', 
                                        'Patricia Garcia',
                                        'Jessica Hernandez'],
                               'Age' : [26, 22 ,24, 21, 25],
                            'Gender' : ['M', 'F', 'M', 'F', 'F'],
                            'Department' : ['ECE', 'EE' ,'DEF', 'XY', 'XYZ'],
                        })
students
```

```python
professors = pd.DataFrame({'Prof_name' : ['Thomas Smith',
                                          'Margaret Garcia',
                                          'James Johnson', 
                                          'Richard Miller',
                                          'Margaret MOORE'],
                     'id_number' : [204, 264, 108, 238, 136],
                     'Department' : ['ECE', 'EE' ,'CSE', 'ME', 'CE'],
                         })
professors
```

```python
lst1 = [10,20,30,40]
lst2 = [1,2,3, 40]
print(lst1+lst2)
```

```python
pd.concat((students, professors)).reset_index().drop(columns = 'index')
```

```python
# For doing merge at least one column should be common 

pd.merge(students, professors)   # Default = inner
```

```python
pd.merge(students, professors, how = 'outer')
```

```python
pd.merge(students, professors, how = 'right')
```

```python
pd.merge(students, professors, how = 'left')
```

```python
pd.merge(students, professors, on = ['id_number', 'Department'])
```

# Iterate over the dataframe

```python
students = pd.DataFrame({'id_number' : [204, 202, 264, 136, 238, 291],
                               'Name' : ['James Smith',
                                        'Maria Rodriguez',
                                        'David Smith', 
                                        'Patricia Garcia',
                                        'Jessica Hernandez',
                                        'James Smith'],
                               'Age' : [26, 22 ,24, 21, 25, 29],
                            'Gender' : ['M', 'F', 'M', 'F', 'F', 'M'],
                            'Department' : ['ECE', 'EE' ,'DEF', 'XY', 'XYZ', 'ME'],
                        })
students
```

```python
for i in students:
    print(i)
```

```python
# Iterating over rows
for i in students.iterrows():
    print('index -->', i[0])
    print(f'value at index {i[0]} -->\n{i[1]}')
    print()
```

```python
# Iterating over column
for i in students.iteritems():
    print('Column -->', i[0])
    print(f'value at column {i[0]} -->\n{i[1]}')
    print()
```

```python
for i in students.Name:
    print(i)
```

```python
students.drop_duplicates(subset = 'Name', keep='first', inplace=False)
```

# Grouping of dataframe

```python
bike_data = pd.read_csv('datasets/citibike_tripdata.csv')
bike_data.head()
```

```python
bike_data['Cust_type'] = np.NaN
bike_data
```

```python
bike_data['Cust_type'] = list(map(lambda val:'Silver' if (val<=500) else 'Gold' if (val>500 and val<=1000) else 'Platinum', bike_data.tripduration))
```

```python
bike_data
```

```python
# Grouping the data based on customer type

list(bike_data.groupby('Cust_type'))
```

```python
bike_data.groupby('Cust_type').get_group('Silver')
```

```python
bike_data.groupby('Cust_type').max()
```

```python
bike_data.groupby('Cust_type').min()
```

```python
bike_data.groupby('Cust_type').mean()
```

```python
bike_data.groupby('Cust_type').describe()
```

```python
bike_data[bike_data['starttime']>'2018-05-01 05:05:46']
```
