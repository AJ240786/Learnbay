```python
# For a given list of values in descending order, 
# write a method in python to search for a value 
# with the help of Binary Search method. 
# The method should return position of the value 
# and should return -1 if the value not present in the list.

lst = [900, 700, 100, 78, 29, 15, 10, 9]

def bin_search(lst, ele):
    f = 0
    l = len(lst)-1
    while f <= l:
        mid = (f+l)//2
        if lst[mid] == ele:
            return mid
        elif lst[mid]>ele:
            f = mid+1
        else:
            l = mid-1
    return -1

bin_search(lst, 15)

# f = 0
# l = 7
# mid = 3
```

```python
my_list = [10,20,30,40,50,60,70,80]
new_lst=[]
if len(my_list)%2==0:
    for i in range(len(my_list)//2,len(my_list)):
        new_lst.append(my_list[i])  #appending second half of my_list as first half of new_list
    for j in range(len(my_list)//2):
        new_lst.append(my_list[j])   #appending first half of my_list as second half of new_list
print(new_lst)
else:
    for i in range(len(my_list)//2+1,len(my_list)):
        new_lst.append(my_list[i])  #appending second half of my_list as first half of new_list
        new_lst.append(my_list[len(my_list)//2])
    for j in range(len(my_list)//2):
        new_lst.append(my_list[j])   #appending first half of my_list as second half of new_list
print(new_lst)
```
