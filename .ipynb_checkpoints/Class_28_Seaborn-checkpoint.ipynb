```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr1 = np.array([1,2, 3, 4, 5, 6, 7, 8])
arr1

```

```python
arr2 = np.array(np.log(arr1))
arr2
```

```python
fig, ax = plt.subplots()

ax.plot(arr1, arr2)

for i in range(len(arr1)):
    plt.annotate( 
                xy = (arr1[i], arr2[i]))
```

# Seaborn

```python
pokemon_data = pd.read_csv('datasets/Pokemon.csv')
pokemon_data.head()
# pokemon_data.shape
pokemon_data['Attack Point'].value_counts().sort_index()
```

```python
pokemon_data.head()
```

```python
# distplot() - distribution plot of the data

sns.distplot(pokemon_data['Attack Point'])
```

```python
hist = sns.distplot(pokemon_data['Attack Point'], kde = False)
plt.show()
```

```python
kde = sns.distplot(pokemon_data['Attack Point'], hist = False, color = 'red')
kde.set_xlabel('Attack Point')
kde.set_ylabel('Density Plot')
kde.set_title('Attack Point vs Density Plot')
plt.show()
```

```python
dist = sns.distplot(pokemon_data['Attack Point'], rug = True)
plt.show()
```

```python
# rugplot()
sns.rugplot(pokemon_data['Attack Point'], 
            height = 0.5, 
            axis = 'x')
```

```python
fig, ax = plt.subplots(figsize = (20, 5))
ax.set_xticks(pokemon_data['Attack Point'])
sns.distplot(pokemon_data['Attack Point'], kde = False, bins = 80)
# for i in range(len(pokemon_data['Attack Point'])):
#     plt.annotate(f'{[i]}, {arr2[i]}', 
#                 xy = (arr1[i], arr2[i]))
```

```python
# Finding the distribution using jointplot() - bivariate analysis

sns.jointplot(x = 'Attack Point',
              y = 'Total Point', 
              data = pokemon_data)
```

```python
sns.jointplot(x = 'Attack Point',
              y = 'Total Point', 
              data = pokemon_data, 
              kind = 'hex')
```

```python
sns.jointplot(x = 'Attack Point',
              y = 'Total Point', 
              data = pokemon_data, 
              kind = 'kde')
```

```python
# heatmap

cormat = pokemon_data.corr()
cormat

```

```python
plt.subplots(figsize = (10, 8))
sns.heatmap(cormat, vmax = 0.8, annot= True, fmt = '.2f', cmap = 'viridis')
```

# Categorical Plot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# swarmplot, boxplot, violinplot, pointplot, countplot, catplot

# At least one value should be a categorical data
```

```python
plt.subplots(figsize = (15, 6))
sns.swarmplot(x = 'day', 
              y = 'tip',
              data = tips)
```

```python
plt.subplots(figsize = (15, 6))
sns.swarmplot(x = 'day', 
              y = 'tip',
              data = tips, 
              hue = 'size')
```

```python
plt.subplots(figsize = (15, 6))
sns.boxplot(x = 'day', 
            y = 'tip',
            data = tips, 
            )

# Whiskers, Inter quartile range, outliers
```

```python
plt.subplots(figsize = (15, 6))
sns.violinplot(x = 'day', 
            y = 'tip',
            data = tips,)
```

```python
plt.subplots(figsize = (15,5))
sns.violinplot(x ='day',
               y = 'tip',
               data = tips,
               inner = None)
sns.swarmplot(x ='day',
              y ='tip',
              data = tips,
              color ='w')
plt.show()
```

```python
sns.countplot(x = 'sex', 
              data = tips)
```

```python
f,ax=plt.subplots(figsize=(15,5))
sns.pointplot(x ='day',
              y ='tip',
              data = tips)
plt.show()
```

```python
sns.catplot(x ='day',y ='tip',data = tips, kind = 'box', col = 'size')
```

```python
sns.catplot(x ='day',y ='tip',data = tips, kind = 'violin')
```
