```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
total_Duration = np.array([117,78,213, 213,207,207,216,219,202,224,238,213,214])
total_Duration
```

```python
krishna_attendance = np.array([117,74,210,210,4,4,209,217,202,223,238,207,213])
krishna_attendance
```

# Basic Plots

```python
plt.plot(total_Duration, krishna_attendance)
plt.xlabel('total_Duration')
plt.ylabel('krishna_attendance')
plt.title('krishna_attendance vs total_Duration')
```

# Generating multiple plots on same canvas


```python
plt.subplot(2, 2, 1)
plt.plot(total_Duration, krishna_attendance, 'r--')
plt.subplot(2, 2, 2)
plt.plot(krishna_attendance, total_Duration,  'b-.')
plt.subplot(2, 2, 3)
plt.plot(krishna_attendance, total_Duration,  'b-.')
plt.subplot(2, 2, 4)
plt.plot(krishna_attendance, total_Duration,  'b-.')
```

# object Oriented method

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(total_Duration, krishna_attendance)
ax.set_xlabel('total_Duration')
ax.set_ylabel('krishna_attendance')
ax.set_title('krishna_attendance vs total_Duration')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.4, 0.2, 0.3, 0.3])

ax1.plot(total_Duration, krishna_attendance)
ax1.set_xlabel('total_Duration')
ax1.set_ylabel('krishna_attendance')
ax1.set_title('krishna_attendance vs total_Duration')

ax2.plot(krishna_attendance, total_Duration)
ax2.set_xlabel('krishna_attendance')
ax2.set_ylabel('total_Duration')
ax2.set_title('total_Duration vs krishna_attendance')
```

```python
# Creating multiple plots using object oriented approach
# subplots() - is an automatic axes manager

fig, ax = plt.subplots()

ax.plot(total_Duration, krishna_attendance)
ax.set_xlabel('total_Duration')
ax.set_ylabel('krishna_attendance')
ax.set_title('krishna_attendance vs total_Duration')
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)

ax[0][0].plot(total_Duration, krishna_attendance)
ax[0][0].set_xlabel('total_Duration')
ax[0][0].set_ylabel('krishna_attendance')
ax[0][0].set_title('krishna_attendance vs total_Duration')

ax[0][1].plot(total_Duration, krishna_attendance)
ax[0][1].set_xlabel('total_Duration')
ax[0][1].set_ylabel('krishna_attendance')
ax[0][1].set_title('krishna_attendance vs total_Duration')

plt.tight_layout()
```

```python
# Figure size and DPI
# figsize = (width, height)


fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 8), dpi = 100)

ax[0][0].plot(total_Duration, krishna_attendance)
ax[0][0].set_xlabel('total_Duration')
ax[0][0].set_ylabel('krishna_attendance')
ax[0][0].set_title('krishna_attendance vs total_Duration')

ax[0][1].plot(total_Duration, krishna_attendance)
ax[0][1].set_xlabel('total_Duration')
ax[0][1].set_ylabel('krishna_attendance')
ax[0][1].set_title('krishna_attendance vs total_Duration')

plt.tight_layout()
```

```python
# Saving the plot
fig.savefig('my_plot.png')
```

# Setting the legends in the graph

```python
fig, ax = plt.subplots()

ax.plot(total_Duration, total_Duration+2, label = 'total_Duration+2')
ax.set_xlabel('total_Duration')
ax.set_ylabel('total_Duration+2')

ax.plot(total_Duration, total_Duration+20, label = 'total_Duration+20')
ax.set_xlabel('total_Duration')
ax.set_ylabel('total_Duration+20')

ax.legend(loc = 0)
```

# Setting the color, linestyle, linewidths, marker

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(total_Duration, total_Duration+2, color = 'b', linestyle = 'solid', linewidth = 0.3)
ax.plot(total_Duration, total_Duration+20, color = 'c', linestyle = 'dashdot', linewidth = 1)
ax.plot(total_Duration, total_Duration+40, color = 'm', linestyle = (0, (1, 10)), linewidth = 5)
ax.plot(total_Duration, total_Duration+60, color = 'r', linestyle = (0, (3, 1, 1, 1, 1, 1)), linewidth = 10)
ax.plot(total_Duration, total_Duration+80, color = 'g', linestyle = (0, (3, 10, 1, 10, 1, 10)), linewidth = 30)
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(total_Duration, total_Duration+2, marker = 'o', markersize = 10, markerfacecolor = 'yellow',
       markeredgecolor = 'red', markeredgewidth = 3)
```

```python
# Setting tickes along x-axis and y-axis
fig, ax = plt.subplots()

ax.plot(total_Duration, krishna_attendance)
ax.set_xticks([0, 50, 100, 150, 200, 250])
ax.set_yticks(range(0, 251, 50))
ax.set_xlabel('total_Duration')
ax.set_ylabel('krishna_attendance')
ax.set_title('krishna_attendance vs total_Duration')
```
